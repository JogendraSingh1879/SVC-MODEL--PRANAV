# Gamma is a parameter in a Support Vector Machine (SVM) that controls how far the influence of a single training example reaches:
## High gamma
Only the closest points to the decision boundary are weighted, resulting in a smoother boundary.
## Low gamma
Even points far from the decision boundary are weighted, resulting in a more wiggly boundary.
Gamma is a hyperparameter that you set before training the model. It's a parameter for the RBF kernel in SVM.
The performance of SVM is affected by the settings of parameters, which can have a profound impact on the resulting engine's implementation.

# In Support Vector Machines (SVM), the parameter **C controls the trade-off between maximizing the margin and minimizing the classification error**
## Smaller C
Emphasizes maximizing the margin, allowing for more misclassifications, and promoting a wider margin. This can be useful for noisy or overlapping data points.
## Larger C
Emphasizes classifying all training examples correctly, even if it means sacrificing the margin. This results in a smaller margin hyperplane, also known as Hard Margin.
## C approaching positive infinity
Results in heavy overfitting as the model pushes for the best possible accuracy.
## C is also known as the penalty parameter or cost parameter for misclassification.
